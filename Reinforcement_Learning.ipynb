```python
import gym
env = gym.make("FrozenLake-v1")


for t in range(1000):
    observation = env.reset()
    env.render()

    for i in range(50):
        action = env.action_space.sample() # random direct
        observation, reward, done, info = env.step(action)
        env.render()
        print(reward, done, info)
        if done:
            break
env.close()     
        
```
